```python
# max cycle per engine
max_cycles = df.groupby("engine_id")["cycle"].max()

# merge back
df = df.merge(
    max_cycles.rename("max_cycle"),
    on="engine_id"
)

# define normal threshold (first 30%)
df["health_ratio"] = df["cycle"] / df["max_cycle"]

# mark normal data
df["is_normal"] = df["health_ratio"] <= 0.3

df.head()
```

```python
df["is_normal"].value_counts()
```

```python
sensor_cols=[col for col in df.columns if col.startswith("sensor")]
df[sensor_cols].std().sort_values()
```

```python
low_var_sensors = df[sensor_cols].std()[df[sensor_cols].std() < 1e-3].index.tolist()

low_var_sensors
```

```python
df_reduced = df.drop(columns=low_var_sensors)
```

```python
from sklearn.preprocessing import StandardScaler
sensor_cols_reduced=[c for c in df_reduced.columns if c.startswith("sensor_")]
scaler = StandardScaler()
df_reduced[sensor_cols_reduced] = scaler.fit_transform(df_reduced[sensor_cols_reduced])
```
